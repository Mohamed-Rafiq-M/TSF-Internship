Visualizing - **Ship Mode**

```
sns.pairplot(data, hue = 'Ship Mode', palette = 'Set1')
plt.show()
```

```
plt.figure(figsize = (15, 10))
plt.subplot(2, 2, 1)
data['Ship Mode'].value_counts().plot.pie(autopct = '%.1f%%')
plt.subplot(2, 2, 2)
sns.countplot(x = 'Ship Mode', data = data, palette = 'Set1')
plt.show()
```

Visualizing - **Sales, Profit and loss across Segment**

```
data.groupby('Ship Mode')[['Sales','Profit']].sum().sort_values(by = 'Sales', ascending = False).plot.bar(figsize = (10, 5))
plt.title('Sales, Profit & loss across the Segments')
plt.xlabel('Ship Mode')
plt.ylabel('Profit, loss & Sales')
plt.show()
```

Visualizing - **Sales, Profit and loss across Ship Mode**

```
data.groupby('Segment')[['Sales','Profit']].sum().sort_values(by = 'Sales', ascending = False).plot.bar(figsize = (10, 5))
plt.title('Sales, Profit & loss across the Ship Mode')
plt.xlabel('Segment')
plt.ylabel('Profit, loss & Sales')
plt.show()
```

Visualizing - **Sales, Profit and loss across Region**

```
data.groupby('Region')[['Sales','Profit']].sum().sort_values(by = 'Sales', ascending = False).plot.bar(figsize = (10, 5))
plt.title('Sales, Profit & loss across the Region')
plt.xlabel('Region')
plt.ylabel('Profit, loss & Sales')
plt.show()
```

Visualizing - **Sales, Profit and loss across Category**

```
data.groupby('Category')[['Sales','Profit']].sum().sort_values(by = 'Sales', ascending = False).plot.bar(figsize = (10, 5))
plt.title('Sales, Profit & loss across the Category')
plt.xlabel('Category')
plt.ylabel('Profit, loss & Sales')
plt.show()
```

Visualizing - **Sales, Profit and loss across Sub-Category**

```
data.groupby('Sub-Category')[['Sales','Profit']].sum().sort_values(by = 'Sales', ascending = False).plot.bar(figsize = (15, 5))
plt.title('Sales, Profit & loss across the Sub-Category')
plt.xlabel('Sub-Category')
plt.ylabel('Profit, loss & Sales')
plt.show()
```

Insights:
* **Phones** and **Chairs** has the highest Sales.
* **Copiers** has the highest profit
* **Tables, Book Cases and Supplies** has no profit and going in loss.





Visualizing - **Sales, Profit and loss across the states**

```
data.groupby('State')[['Sales','Profit']].sum().sort_values(by = 'Sales',ascending = False).plot.bar(figsize = (23, 10))
plt.title('Sales, Profit & loss across the States')
plt.xlabel('States')
plt.ylabel('Profit, loss & Sales')
plt.show()
```

Insights:
* **California** state has the highest sales and highest profit over all other states.
* **North Dakota** state has the least sales over all other states.
* **Wyoming** state has the least profit over all other states.
* **Texas, Pennsylvania, Florida, Illinois, Ohio, North Carolina, Arizona, Oregon, Colorado, Tennessee** states has no profit and it is in loss.
* **Texas** state has the highest loss and **Oregon** state has the compared to the other states which are in loss.

Overall Observations:
* The **Standard Class** ship mode was preferred highest whereas **Same Day** ship mode was preferred least among all the ship modes.
* The majority of buyers were **Consumer**.
* People residing in **West** region were major buyers among all the regions.
* The **Office Supplies** were the most preferred category by the buyers.
* **Binders** and **Paper** were preferred most and **Copiers** and **Machines** were prefered least among all the sub-category.
* **Phones** has the highest sales and **Copiers** has highest profit among the sub-category. **Tabels** has highest loss compared to all.
* **California** state has highest sales and profit compared to all other states. **Texas** state has highest loss compared to all other states that were in loss.
